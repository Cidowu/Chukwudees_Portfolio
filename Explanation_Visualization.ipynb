# **Ford GoBike System User Anlysis**

## **Investigation Overview**

In this investigation, I wanted to look at the characteristics of the Ford GoBike system users in the the greater San Francisco Bay area. I wanted to understand their behaviour in terms of trevel time by daily and weekly travels, and other characteristics like age, gender, and user type.


## **Dataset Overview**
After cleaning, There were 174,760 riders' information over a period of one month in the dataset with 11 features. The features listed below are the features that were of interest to me.

*  rider age
*  user type
*   member gender
*   start time of day
*   week


```python
#import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
#import data
path = "fordGoBike_cleaned.csv"
ride_df = pd.read_csv(path)
```

### **Distribution of user type**

The user type are primarily made up of two groups,the Subscribers, and the Customers. The visualization below show that the Subsribers make up a large portion of the market, while the Customers make up a smaller portion. 

```python
plt.figure(figsize = (5,5))
sns.countplot(data = ride_df, x = 'user_type', order = ride_df['user_type'].value_counts().index, color = sns.color_palette()[0]);
plt.title('Distribution of Users');
```

### **User Age Distribution**

The data agrees with logical reasoning, and shows that the distribution of the market is skewed to the right, with the majority of the users being between 20 and 40 years old

```python
plt.figure(figsize=(6,6))
plt.hist(ride_df['rider_age'],bins = 15);
plt.title('User Age Distribution');
plt.xlabel('User Age (Years)');
plt.ylabel('counts');
```

### **User Gender Distribution**

The bar chart below show the gender distribution of the users of the Ford GoBike system. The visulization below was created to provide a clearer idea into how interested the different genders are engaged in bike ride sharing. It appears that more males engage in bike riding when compared to the females, and the other gender.

```python
fig, ax = plt.subplots(figsize = [5, 5])
default_color = sns.color_palette()[0]
sns.countplot(data = ride_df, x = 'member_gender', color = default_color, ax = ax)
plt.title('User Gender Distribution');
plt.show()
```

### **User Gender vs. User Type**

Generally, the user type distribution across all genders is the same. The subcribers are predominant amongst both categories



```python
plt.figure(figsize = [5, 5]);
sns.countplot(data = ride_df, x = 'member_gender', hue = 'user_type');
plt.title('USER GENDER DISTRIBUTION WITH USER TYPE');
```

### **Start time of day vs. Member gender vs. User Type vs. Day of the week**

The peak times for travel is mornings, evenings, and afternoons, in that order. This distribution was also oberved across genders and user types, with a little exception to customers. Customers tend to use the product mostly in the mornings and afternoons, especially in the afternoon

```python
plt.figure(figsize = [10, 10])

# subplot 1: Start time of day vs. Week
plt.subplot(3, 1, 1)
sns.countplot(data = ride_df, x = 'start_time_of_day', hue = 'week', palette = 'Blues')
plt.title("Exploring the relationship between Start time of day, Member gender, User Type, and Day of the week")

# subplot 2: Start time of day vs. Gender
ax = plt.subplot(3, 1, 2)
sns.countplot(data = ride_df, x = 'start_time_of_day', hue = 'member_gender', palette = 'Reds')
ax.legend(ncol = 2) 

# subplot 3: Start time of day vs. User Type
ax = plt.subplot(3, 1, 3)
sns.countplot(data = ride_df, x = 'start_time_of_day', hue = 'user_type', palette = 'Greens')
ax.legend(loc = 1, ncol = 2) 
plt.show()
```

### **Conclusion**
