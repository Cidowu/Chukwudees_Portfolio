### **Introuction**

Data wrangling is the process of gathering data from different sources, in different formats, for the purpose of drawing analytical insights from the data, and using the insights to understand patterns and make decisions. Data wrangling is done in stages, and include, gathering, assessing, cleaning, and storage.

This data wrangling project involves data gathering from the "weRateDogs" twitter archive, and the twitter API with "tweepy". Following the data gathering process, the assessment of the data to identify quality and tidiness issues with the data was done both visually and programatically. The issues identified in the assessment were further cleaned in-line with the "Define-Code-Test" framework, before the cleaned data were then stored in "csv" formats.
